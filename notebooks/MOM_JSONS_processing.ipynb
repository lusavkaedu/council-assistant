```python
from typing import Optional, List
from datetime import date

class Person:
    def __init__(self, name: str, role: Optional[str] = None, party: Optional[str] = None,
                 type_: str = "councillor", known_titles: Optional[List[str]] = None):
        self.name = name
        self.role = role
        self.party = party
        self.type = type_
        self.known_titles = known_titles or []
        self.attendance_log = []

    def log_attendance(self, meeting_id: str, date_: date, status: str):
        self.attendance_log.append((meeting_id, date_, status))

    def attendance_summary(self, start_date: date, end_date: date):
        relevant = [s for mid, dt, s in self.attendance_log if start_date <= dt <= end_date]
        total = len(relevant)
        present = relevant.count("present")
        return {
            "total_meetings": total,
            "present": present,
            "percentage": (present / total * 100) if total else 0.0
        }

    def __repr__(self):
        return f"<Person: {self.name} ({self.type})>"
```

```python
import pandas as pd
from datetime import datetime

# Use your JSONs
docs = [doc_1, doc_2, doc_3]

# Step 1: Flatten attendance records
attendance_records = []
for doc in docs:
    meeting_date = datetime.strptime(doc["meeting_date"], "%Y-%m-%d")
    committee = doc["committee"]
    for status in ['present', 'absent', 'virtual']:
        for person in doc.get("attendance", {}).get(status, []):
            name_clean = person.split(" (")[0].strip()
            attendance_records.append({
                "Person": name_clean,
                "Meeting ID": doc["meeting_id"],
                "Date": meeting_date,
                "Committee": committee,
                "Status": status
            })

df_attendance = pd.DataFrame(attendance_records)

# Step 2: Add Month, and flag 'Present'
df_attendance["Month"] = df_attendance["Date"].dt.to_period("M").astype(str)
df_attendance["Present"] = df_attendance["Status"] == "present"

# Step 3: Create full grid of committees × months
all_committees = sorted(set(df_attendance["Committee"]))
all_months = pd.period_range("2024-01", "2024-12", freq="M").astype(str)
index = pd.MultiIndex.from_product([all_committees, all_months], names=["Committee", "Month"])
full_grid = pd.DataFrame(index=index).reset_index()

# Step 4: Filter for one person
def enriched_attendance_matrix(person_name):
    # Filter attendance for person
    df_person = df_attendance[df_attendance["Person"] == person_name][["Committee", "Month", "Present"]].drop_duplicates()

    # Merge with full grid
    merged = full_grid.merge(df_person, on=["Committee", "Month"], how="left")
    merged["Present"] = merged["Present"].where(merged["Present"].notnull(), "N/A")

    # Pivot
    matrix = merged.pivot(index="Committee", columns="Month", values="Present")

    # Compute % Present
    numeric_matrix = matrix.copy()
    for col in matrix.columns:
        numeric_matrix[col] = matrix[col].apply(
            lambda x: True if x is True else (False if x is False else pd.NA)
        )
    matrix["% Present"] = numeric_matrix.apply(
        lambda row: (row == True).sum() / row.count() * 100 if row.count() else 0,
        axis=1
    )

    # Apply styling
    def highlight(val):
        if val is True:
            return "background-color: lightgreen"
        elif val is False:
            return "background-color: lightcoral"
        elif val == "N/A":
            return "color: gray"
        return ""

    styled = matrix.style.applymap(highlight, subset=pd.IndexSlice[:, matrix.columns[:-1]]).format({"% Present": "{:.1f}%"})
    return styled

# Step 5: Display for a chosen councillor
enriched_attendance_matrix("Mr P Cole")
```

```python
import re
from datetime import datetime

# === Load text ===
#with open("may_2025_questions.txt", "r") as f:
#    text_raw = f.read()

# === Clean page and metadata artifacts ===
cleaned_text = re.sub(r"Page\s*\d+Agenda Item.*", "", text_raw)
cleaned_text = re.sub(r"Page\s*\d+", "", cleaned_text)
cleaned_text = re.sub(r"This page is intentionally left blank", "", cleaned_text, flags=re.I)
cleaned_text = re.sub(r"\n{3,}", "\n\n", cleaned_text)  # collapse excessive spacing

# === Split into Q blocks ===
qa_blocks = re.split(r"\bQuestion\s+(\d+)\b", cleaned_text)[1:]  # retain question numbers
parsed_qas = []

for i in range(0, len(qa_blocks), 2):
    q_number = int(qa_blocks[i])
    block = qa_blocks[i + 1].strip()

    # Skip rejected questions
    if "Question rejected" in block:
        continue

    # Extract date
    date_match = re.search(r"COUNTY COUNCIL\s*\n\s*Thursday\s+(\d{1,2}\s+\w+\s+\d{4})", block)
    date_str = date_match.group(1).strip() if date_match else None
    date_iso = datetime.strptime(date_str, "%d %B %Y").date().isoformat() if date_str else None

    # Extract speaker metadata
    q_meta_match = re.search(r"Question by (.+?) to (.+?)\n", block)
    if q_meta_match:
        councillor = q_meta_match.group(1).strip()
        target = q_meta_match.group(2).strip()
    else:
        councillor = target = None

    # Split on 'Answer'
    q_split = re.split(r"\nAnswer\n", block, maxsplit=1)
    question_text = q_split[0].strip()
    answer_text = q_split[1].strip() if len(q_split) > 1 else ""

    # Clean final junk from content
    def clean_text(t):
        t = re.sub(r"Page\s*\d+.*", "", t)
        return t.strip()

    question_text = clean_text(question_text)
    answer_text = clean_text(answer_text)

    parsed_qas.append({
        "question_number": q_number,
        "meeting_date": date_iso,
        "councillor": councillor,
        "question_to": target,
        "question_text": question_text,
        "answer_text": answer_text
    })

parsed_qas
```

```python
template = """
<!DOCTYPE html>
<html>
<head>
  <meta charset="UTF-8">
  <title>Auto Knowledge Graph</title>
  <script src="https://d3js.org/d3.v7.min.js"></script>
  <style>
    .node circle { fill: #87cefa; stroke: #1f78b4; stroke-width: 2px; cursor: grab; }
    .node text { font-size: 12px; }
    .link { stroke: #999; stroke-opacity: 0.6; stroke-width: 2px; }
    .label { font-size: 11px; fill: #555; pointer-events: none; }
  </style>
</head>
<body>
  <svg width="960" height="720"></svg>
  <script>
    const nodes = {{NODES}};
    const links = {{LINKS}};

    const svg = d3.select("svg"),
          width = +svg.attr("width"),
          height = +svg.attr("height");

    const simulation = d3.forceSimulation(nodes)
      .force("link", d3.forceLink(links).id(d => d.id).distance(170))
      .force("charge", d3.forceManyBody().strength(-500))
      .force("center", d3.forceCenter(width / 2, height / 2));

    const link = svg.append("g").selectAll("line")
      .data(links).join("line").attr("class", "link");

    const edgeLabels = svg.append("g").selectAll(".label")
      .data(links).enter().append("text")
      .attr("class", "label").text(d => d.topic);

    const node = svg.append("g").selectAll("g")
      .data(nodes).join("g").attr("class", "node")
      .call(d3.drag()
        .on("start", dragstarted)
        .on("drag", dragged)
        .on("end", dragended));

    node.append("circle").attr("r", 20);
    node.append("text").text(d => d.id).attr("x", 25).attr("y", 5);

    simulation.on("tick", () => {
      link.attr("x1", d => d.source.x).attr("y1", d => d.source.y)
          .attr("x2", d => d.target.x).attr("y2", d => d.target.y);
      node.attr("transform", d => `translate(${d.x},${d.y})`);
      edgeLabels.attr("x", d => (d.source.x + d.target.x) / 2)
                .attr("y", d => (d.source.y + d.target.y) / 2);
    });

    function dragstarted(event, d) {
      if (!event.active) simulation.alphaTarget(0.3).restart();
      d.fx = d.x; d.fy = d.y;
    }
    function dragged(event, d) {
      d.fx = event.x; d.fy = event.y;
    }
    function dragended(event, d) {
      if (!event.active) simulation.alphaTarget(0);
      d.fx = null; d.fy = null;
    }
  </script>
</body>
</html>
"""

#with open("kcc_graph_template.html", "w") as f:
#    f.write(template)
```

```python
def generate_knowledge_graph_html(text_raw, output_path="auto_graph.html"):
    import json, re

    # Use directly if it's a list, otherwise try to parse
    if isinstance(text_raw, str):
        data = json.loads(text_raw)
    else:
        data = text_raw

    nodes_set = set()
    edges = []

    for item in data:
        src = item["councillor"].strip()
        tgt = item["question_to"].strip()
        text = item["question_text"]
        topic = " ".join(re.findall(r'[A-Za-z]+', text)[-6:])  # crude fallback topic

        nodes_set.add(src)
        nodes_set.add(tgt)
        edges.append({"source": src, "target": tgt, "topic": topic})

    nodes = [{"id": n} for n in sorted(nodes_set)]
    nodes_json = json.dumps(nodes, indent=2)
    edges_json = json.dumps(edges, indent=2)

    with open("kcc_graph_template.html") as f:
        template = f.read()

    html_output = template.replace("{{NODES}}", nodes_json).replace("{{LINKS}}", edges_json)

    with open(output_path, "w") as f:
        f.write(html_output)

    return output_path
```
